```python
stack = []

#append() fungsi untuk memasukkan elemen ke dalam stack
stack.append('1')
stack.append('2')
stack.append('3')

print('Isi stack : ', stack)
print(stack)

#pop() fungsi untuk menghapus elemen dari stack
print('Elemen yang dikeluarkan : ', stack.pop())
print(stack.pop())
print(stack.pop())

print('\nIsi stack setelah dihapus: ', stack)
print(stack)
```

```python
from collections import deque
stack = deque()

#append() fungsi untuk memasukkan elemen ke dalam stack
stack.append('1')
stack.append('2')
stack.append('3')

print('Isi stack : ', stack)
print(stack)

#pop() fungsi untuk menghapus elemen dari stack
print('\nElemen yang dikeluarkan : ', stack.pop())
print(stack.pop())
print(stack.pop())

print('\nIsi stack setelah dihapus: ', stack)
print(stack)

print('\nIsi stack setelah dihapus: ', stack)
print(stack)
```
